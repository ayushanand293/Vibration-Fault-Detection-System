## 1Ô∏è‚É£ Signal Generation

We generate synthetic vibration signals that mimic real CWRU bearing data. Each fault type has a characteristic frequency:

| Fault Type | Frequency | Description |
|-----------|-----------|-------------|
| **Normal** | 30 Hz | Healthy bearing (motor rotation) |
| **Inner Race** | 297 Hz | Inner race defect |
| **Outer Race** | 250 Hz | Outer race defect |
| **Ball** | 400 Hz | Ball bearing defect |

---

```python
def generate_signal(fault_type, duration=2, fs=12000):
    """
    Generate realistic vibration signals
    
    Parameters:
    - fault_type: 'normal', 'inner_race', 'outer_race', 'ball'
    - duration: Signal duration in seconds
    - fs: Sampling frequency (Hz)
    
    Returns:
    - t: Time array
    - s: Signal array
    - fs: Sampling frequency
    """
    t = np.linspace(0, duration, fs * duration, endpoint=False)
    
    if fault_type == 'normal':
        # Healthy bearing: low amplitude, motor frequency
        s = 0.3 * np.sin(2 * np.pi * 30 * t)
        s += np.random.normal(0, 0.02, len(t))
        
    elif fault_type == 'inner_race':
        # Inner race fault: 297 Hz characteristic frequency
        s = 0.5 * np.sin(2 * np.pi * 297 * t)
        s += np.random.normal(0, 0.05, len(t))
        
    elif fault_type == 'outer_race':
        # Outer race fault: 250 Hz
        s = 0.6 * np.sin(2 * np.pi * 250 * t)
        s += np.random.normal(0, 0.03, len(t))
        
    else:  # ball
        # Ball fault: 400 Hz
        s = 0.4 * np.sin(2 * np.pi * 400 * t)
        s += np.random.normal(0, 0.04, len(t))
    
    return t, s, fs

# Generate all signals
fault_types = ['normal', 'inner_race', 'outer_race', 'ball']
signals = {}

print("üîÑ Generating vibration signals...\n")
for ft in fault_types:
    t, s, fs = generate_signal(ft)
    signals[ft] = {'time': t, 'signal': s, 'fs': fs}
    print(f"  ‚úÖ {ft:12s}: {len(s):,} samples @ {fs:,} Hz")

print("\n‚úÖ All signals generated successfully!")
```

```python
# Plot raw signals
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Raw Vibration Signals - Time Domain', fontsize=18, fontweight='bold', y=0.995)

for idx, ft in enumerate(fault_types):
    ax = axes[idx // 2, idx % 2]
    t = signals[ft]['time'][:5000]  # First 5000 samples
    s = signals[ft]['signal'][:5000]
    
    ax.plot(t, s, linewidth=0.8, alpha=0.8)
    ax.set_title(f'{ft.upper().replace("_", " ")}', fontweight='bold', fontsize=14)
    ax.set_xlabel('Time (seconds)', fontsize=11)
    ax.set_ylabel('Amplitude', fontsize=11)
    ax.grid(True, alpha=0.3)
    
    # Add statistics
    rms = np.sqrt(np.mean(s**2))
    peak = np.max(np.abs(s))
    ax.text(0.02, 0.98, f'RMS: {rms:.3f}\nPeak: {peak:.3f}', 
            transform=ax.transAxes, fontsize=10, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.savefig('../reports/raw_signals.png', dpi=300, bbox_inches='tight')
print("‚úÖ Saved: reports/raw_signals.png")
plt.show()
```

## 2Ô∏è‚É£ Preprocessing Pipeline

We apply a 5-stage preprocessing pipeline to clean the raw signals:

### Pipeline Stages:
1. **Detrending** - Remove linear trends
2. **Moving Average Filter** - Smooth high-frequency noise (window=5)
3. **Low-Pass Butterworth Filter** - Remove frequencies > 5000 Hz (order=4)
4. **Hanning Window** - Reduce spectral leakage
5. **Normalization** - Zero mean, unit variance (Z-score)

---

```python
def preprocess_signal(raw, fs):
    """
    Complete preprocessing pipeline
    
    Parameters:
    - raw: Raw signal array
    - fs: Sampling frequency
    
    Returns:
    - normalized: Final processed signal
    - stages: Dictionary of intermediate stages
    """
    # Stage 1: Detrending
    detrended = signal.detrend(raw, type='linear')
    
    # Stage 2: Moving Average Filter (5-point)
    ma = np.convolve(detrended, np.ones(5)/5, mode='same')
    
    # Stage 3: Low-Pass Butterworth Filter (5000 Hz cutoff)
    nyquist = fs / 2
    cutoff = 5000 / nyquist
    b, a = signal.butter(4, cutoff, btype='low')
    filtered = signal.filtfilt(b, a, ma)
    
    # Stage 4: Hanning Window
    window = signal.windows.hann(len(filtered))
    windowed = filtered * window
    
    # Stage 5: Normalization (Z-score)
    normalized = (windowed - np.mean(windowed)) / (np.std(windowed) + 1e-10)
    
    # Store all stages
    stages = {
        'raw': raw,
        'detrended': detrended,
        'ma_filtered': ma,
        'lp_filtered': filtered,
        'windowed': windowed,
        'normalized': normalized
    }
    
    return normalized, stages

# Preprocess all signals
processed_signals = {}

print("üîÑ Preprocessing signals...\n")
for ft in fault_types:
    proc, stages = preprocess_signal(signals[ft]['signal'], signals[ft]['fs'])
    processed_signals[ft] = {'processed': proc, 'stages': stages}
    print(f"  ‚úÖ {ft:12s}: Preprocessed")

print("\n‚úÖ All signals preprocessed!")
```

```python
# Show preprocessing pipeline for NORMAL bearing
fig, axes = plt.subplots(2, 3, figsize=(18, 10))
fig.suptitle('Preprocessing Pipeline - NORMAL Bearing', fontsize=18, fontweight='bold', y=0.995)

stages = processed_signals['normal']['stages']
stage_names = ['Raw Signal', 'Detrended', 'MA Filtered', 
               'LP Filtered', 'Windowed', 'Normalized']
stage_keys = ['raw', 'detrended', 'ma_filtered', 
              'lp_filtered', 'windowed', 'normalized']
colors = ['steelblue', 'coral', 'green', 'purple', 'orange', 'red']

for idx, (name, key, color) in enumerate(zip(stage_names, stage_keys, colors)):
    ax = axes[idx // 3, idx % 3]
    data = stages[key][:5000]  # First 5000 samples
    
    ax.plot(data, linewidth=0.8, color=color, alpha=0.8)
    ax.set_title(f'Stage {idx+1}: {name}', fontweight='bold', fontsize=12)
    ax.set_xlabel('Sample Index', fontsize=10)
    ax.set_ylabel('Amplitude', fontsize=10)
    ax.grid(True, alpha=0.3)
    
    # Statistics
    mean = np.mean(data)
    std = np.std(data)
    ax.text(0.02, 0.98, f'Œº={mean:.3f}\nœÉ={std:.3f}', 
            transform=ax.transAxes, fontsize=9, verticalalignment='top',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

plt.tight_layout()
plt.savefig('../reports/01_preprocessing_pipeline.png', dpi=300, bbox_inches='tight')
print("‚úÖ Saved: reports/01_preprocessing_pipeline.png")
plt.show()
```

## 3Ô∏è‚É£ Frequency-Domain Analysis

### Fast Fourier Transform (FFT)
Converts time-domain signals to frequency-domain to identify characteristic fault frequencies.

### Power Spectral Density (PSD)
Shows power distribution across frequencies using Welch's method.

**Key Insight:** Each fault type has a unique frequency signature!

---

```python
def compute_fft_psd(sig, fs):
    """
    Compute FFT and PSD
    
    Parameters:
    - sig: Preprocessed signal
    - fs: Sampling frequency
    
    Returns:
    - freqs: Frequency array for FFT
    - fft_vals: FFT magnitude values (normalized)
    - freqs_psd: Frequency array for PSD
    - psd: Power spectral density
    """
    n = len(sig)
    
    # FFT
    freqs = np.fft.fftfreq(n, d=1/fs)[:n//2]
    fft_vals = np.abs(np.fft.fft(sig)[:n//2])
    fft_vals = fft_vals / (np.max(fft_vals) + 1e-10)  # Normalize
    
    # PSD using Welch's method
    freqs_psd, psd = signal.welch(sig, fs=fs, nperseg=1024)
    
    return freqs, fft_vals, freqs_psd, psd

# Compute for all signals
frequency_data = {}

print("üîÑ Computing FFT & PSD...\n")
for ft in fault_types:
    proc = processed_signals[ft]['processed']
    freqs, fft, freqs_psd, psd = compute_fft_psd(proc, 12000)
    frequency_data[ft] = {
        'freqs': freqs, 
        'fft': fft, 
        'freqs_psd': freqs_psd, 
        'psd': psd
    }
    
    # Find dominant frequency
    dominant_freq = freqs[np.argmax(fft)]
    print(f"  ‚úÖ {ft:12s}: Dominant frequency = {dominant_freq:.1f} Hz")

print("\n‚úÖ Frequency analysis complete!")
```

```python
# Plot FFT for all fault types
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('FFT Analysis - Frequency Spectra', fontsize=18, fontweight='bold', y=0.995)

for idx, ft in enumerate(fault_types):
    ax = axes[idx // 2, idx % 2]
    freqs = frequency_data[ft]['freqs'][:2000]  # 0-1000 Hz
    fft = frequency_data[ft]['fft'][:2000]
    
    ax.plot(freqs, fft, linewidth=1, alpha=0.8)
    ax.set_title(f'{ft.upper().replace("_", " ")}', fontweight='bold', fontsize=14)
    ax.set_xlabel('Frequency (Hz)', fontsize=11)
    ax.set_ylabel('Normalized Magnitude', fontsize=11)
    ax.grid(True, alpha=0.3)
    
    # Mark peak frequency
    peak_idx = np.argmax(fft)
    peak_freq = freqs[peak_idx]
    ax.axvline(peak_freq, color='red', linestyle='--', linewidth=2, alpha=0.7)
    ax.text(peak_freq + 10, 0.9, f'Peak: {peak_freq:.1f} Hz', 
            fontsize=11, fontweight='bold', color='red')

plt.tight_layout()
plt.savefig('../reports/02_fft_comparison.png', dpi=300, bbox_inches='tight')
print("‚úÖ Saved: reports/02_fft_comparison.png")
plt.show()
```

```python
# Plot PSD for all fault types
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('PSD Analysis - Power Spectral Density', fontsize=18, fontweight='bold', y=0.995)

for idx, ft in enumerate(fault_types):  # ‚Üê FIXED: Added enumerate()
    ax = axes[idx // 2, idx % 2]
    freqs_psd = frequency_data[ft]['freqs_psd']
    psd = frequency_data[ft]['psd']
    
    ax.semilogy(freqs_psd, psd, linewidth=1, alpha=0.8)
    ax.set_title(f'{ft.upper().replace("_", " ")}', fontweight='bold', fontsize=14)
    ax.set_xlabel('Frequency (Hz)', fontsize=11)
    ax.set_ylabel('Power/Frequency (dB/Hz)', fontsize=11)
    ax.grid(True, alpha=0.3)
    
    # Mark peak
    peak_idx = np.argmax(psd)
    peak_freq = freqs_psd[peak_idx]
    ax.axvline(peak_freq, color='red', linestyle='--', linewidth=2, alpha=0.7)

plt.tight_layout()
plt.savefig('../reports/psd_analysis.png', dpi=300, bbox_inches='tight')
print("‚úÖ Saved: reports/psd_analysis.png")
plt.show()
```

## 4Ô∏è‚É£ Feature Extraction

We extract **14 features** from each signal:

### Time-Domain Features (6):
- RMS (Root Mean Square)
- Peak Amplitude
- Crest Factor
- Kurtosis
- Skewness
- Standard Deviation

### Frequency-Domain Features (8):
- Dominant Frequency
- Peak FFT Magnitude
- Top 3 Frequencies
- Spectral Entropy
- Frequency Centroid

---

```python
def extract_features(sig, freqs, fft_vals, freqs_psd, psd):
    """
    Extract 14 features from signal
    """
    features = {}
    
    # === TIME-DOMAIN FEATURES ===
    features['rms'] = np.sqrt(np.mean(sig**2))
    features['peak'] = np.max(np.abs(sig))
    features['peak_to_peak'] = np.max(sig) - np.min(sig)
    features['crest_factor'] = features['peak'] / (features['rms'] + 1e-10)
    features['kurtosis'] = kurtosis(sig)
    features['skewness'] = pd.Series(sig).skew()
    features['std_dev'] = np.std(sig)
    
    # === FREQUENCY-DOMAIN FEATURES ===
    peak_idx = np.argmax(fft_vals)
    features['dominant_frequency'] = freqs[peak_idx]
    features['peak_fft_magnitude'] = fft_vals[peak_idx]
    
    # Top 3 frequencies
    top3_indices = np.argsort(fft_vals)[-3:][::-1]
    features['top_freq_1'] = freqs[top3_indices[0]]
    features['top_freq_2'] = freqs[top3_indices[1]]
    features['top_freq_3'] = freqs[top3_indices[2]]
    
    # Spectral entropy
    normalized_psd = psd / (np.sum(psd) + 1e-10)
    spectral_entropy = -np.sum(normalized_psd * np.log2(normalized_psd + 1e-10))
    features['spectral_entropy'] = spectral_entropy
    
    # Frequency centroid
    freq_centroid = np.sum(freqs_psd * psd) / (np.sum(psd) + 1e-10)
    features['frequency_centroid'] = freq_centroid
    
    return features

# Extract features for all signals
all_features = []

print("üîÑ Extracting features...\n")
for ft in fault_types:
    proc = processed_signals[ft]['processed']
    freqs = frequency_data[ft]['freqs']
    fft = frequency_data[ft]['fft']
    freqs_psd = frequency_data[ft]['freqs_psd']
    psd = frequency_data[ft]['psd']
    
    features = extract_features(proc, freqs, fft, freqs_psd, psd)
    features['fault_type'] = ft
    all_features.append(features)
    
    print(f"  ‚úÖ {ft:12s}: 14 features extracted")

# Create DataFrame
df_features = pd.DataFrame(all_features)
print("\n‚úÖ Feature extraction complete!")
print(f"\nüìä Dataset shape: {df_features.shape}")
```

```python
# Visualize key features
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
fig.suptitle('Feature Comparison Across Fault Types', fontsize=18, fontweight='bold', y=0.995)

features_to_plot = ['rms', 'kurtosis', 'crest_factor', 'dominant_frequency']
titles = ['RMS (Root Mean Square)', 'Kurtosis', 'Crest Factor', 'Dominant Frequency (Hz)']

for idx, (feat, title) in enumerate(zip(features_to_plot, titles)):
    ax = axes[idx // 2, idx % 2]
    
    # Get values for each fault type
    values = [df_features[df_features['fault_type'] == ft][feat].values[0] 
              for ft in fault_types]
    colors = ['green', 'orange', 'red', 'purple']
    
    # Create bar chart
    bars = ax.bar(range(len(fault_types)), values, color=colors, alpha=0.7, edgecolor='black', linewidth=2)
    
    # Configure axes
    ax.set_xticks(range(len(fault_types)))
    ax.set_xticklabels([ft.replace('_', '\n').title() for ft in fault_types], fontsize=10)
    ax.set_title(title, fontweight='bold', fontsize=13)
    ax.set_ylabel('Value', fontsize=11)
    ax.grid(True, alpha=0.3, axis='y')
    
    # Add value labels on top of bars
    for bar, val in zip(bars, values):
        height = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2., height,
                f'{val:.2f}', ha='center', va='bottom', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('../reports/03_feature_comparison.png', dpi=300, bbox_inches='tight')
print("‚úÖ Saved: reports/03_feature_comparison.png")
plt.show()
```
